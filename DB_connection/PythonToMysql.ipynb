```python
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/purchase_details") # reconnecting to db
```

```python
with engine.connect() as conn:
    conn.execute(text("create table if not exists location(location_id varchar(5) PRIMARY KEY,location_name varchar(30))"))
    conn.commit()
```

```python
with engine.connect() as conn:
    conn.execute(text(""" 
    create table if not exists shipping_info (
    shipping_id VARCHAR(8) PRIMARY KEY,
    shipping_type varchar(15))
    """))
    conn.commit()
```

```python
with engine.connect() as conn:
    conn.execute(text(""" 
    create table if not exists payment_info (
    payment_id VARCHAR(8) PRIMARY KEY,
    payment_type varchar(20))
    """))
    conn.commit()
```

```python
with engine.connect() as conn:
    conn.execute(text(""" 
    create table if not exists discount_info (
    promo_id varchar(8) PRIMARY KEY,
    promo_code varchar(10),
    discount_applied varchar(3))
    """))
    conn.commit()
```

```python
with engine.connect() as conn:
    conn.execute(text(""" 
    create table if not exists product_info (
    product_id VARCHAR(10) PRIMARY KEY,
    category varchar(15),
    item_purchased varchar(15),
    size varchar(3),
    color varchar(10))
    """))
    conn.commit()
```

```python
with engine.connect() as conn:
    conn.execute(text(""" 
    create table if not exists customer_info (
    customer_id int PRIMARY KEY,
    age int,
    gender varchar(10),
    season varchar(10),
    frequency_of_purchased int,
    subscription varchar(3))
    """))
    conn.commit()
```

```python
with engine.connect() as conn:
    conn.execute(text("""
    CREATE TABLE IF NOT EXISTS purchase_info (
        purchase_id INT PRIMARY KEY AUTO_INCREMENT,
        customer_id INT,
        product_id VARCHAR(10),
        location_id varchar(5),
        payment_id varchar(8),
        promo_id varchar(8),
        shipping_id VARCHAR(8),
        purchase_amount_usd DECIMAL(10,2),
        review_rating DECIMAL(2,1),
        previous_purchases INT,

        FOREIGN KEY (customer_id) REFERENCES customer_info(customer_id),
        FOREIGN KEY (product_id) REFERENCES product_info(product_id),
        FOREIGN KEY (location_id) REFERENCES location(location_id),
        FOREIGN KEY (payment_id) REFERENCES payment_info(payment_id),
        FOREIGN KEY (promo_id) REFERENCES discount_info(promo_id),
        FOREIGN KEY (shipping_id) REFERENCES shipping_info(shipping_id)
    )
    """))
    conn.commit()
```
